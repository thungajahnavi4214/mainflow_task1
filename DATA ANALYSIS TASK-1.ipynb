```python
# Importing necessary libraries
# I am importing pandas, numpy for data manipulation and matplotlib for visualization
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Data Loading
# Reading the dataset using pandas
# I have assumed the file is in the same directory; adjust the file path as needed
# The dataset has a semicolon as the delimiter, so delimiter is specified explicitly
data = pd.read_csv(r"C:\Users\I love daddy\Documents\student-mat.csv", delimiter=';')

# Displaying the first few rows to understand the dataset structure
# This helps to verify the dataset is loaded correctly
print("Dataset Preview:")
print(data.head())

# Step 2: Data Exploration
# Checking for missing values in the dataset
# I want to know if any column has missing values
missing_values = data.isnull().sum()
print("\nMissing Values per Column:")
print(missing_values)

# Displaying the data types of each column
# This is essential to ensure compatibility for calculations
data_types = data.dtypes
print("\nData Types:")
print(data_types)

# Understanding the size of the dataset
# Checking the number of rows and columns
data_shape = data.shape
print("\nDataset Shape (Rows, Columns):")
print(data_shape)

# Step 3: Data Cleaning
# Handling missing values by replacing them with the median (if any exist)
# Median is chosen to minimize the effect of outliers
if missing_values.sum() > 0:
    data = data.fillna(data.median(numeric_only=True))
    print("\nMissing values handled by replacing with median.")

# Removing duplicate rows to ensure clean data
data = data.drop_duplicates()
print("\nDuplicates removed. New dataset shape:", data.shape)

# Step 4: Data Analysis Questions
# Question 1: What is the average score in math (G3)?
average_score = data['G3'].mean()
print("\nAverage Final Grade (G3):", average_score)

# Question 2: How many students scored above 15 in their final grade (G3)?
students_above_15 = (data['G3'] > 15).sum()
print("\nNumber of Students Scoring Above 15 in G3:", students_above_15)

# Question 3: Is there a correlation between study time and final grade (G3)?
studytime_correlation = data['studytime'].corr(data['G3'])
print("\nCorrelation between Study Time and G3:", studytime_correlation)

# Question 4: Which gender has a higher average final grade (G3)?
gender_avg_scores = data.groupby('sex')['G3'].mean()
print("\nAverage Final Grade by Gender:")
print(gender_avg_scores)

# Step 5: Data Visualization
# Plot 1: Histogram of Final Grades (G3)
plt.figure(figsize=(8, 5))
plt.hist(data['G3'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Final Grades (G3)')
plt.xlabel('Final Grade (G3)')
plt.ylabel('Number of Students')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Plot 2: Scatter Plot (Study Time vs. Final Grade)
plt.figure(figsize=(8, 5))
plt.scatter(data['studytime'], data['G3'], color='purple', alpha=0.7)
plt.title('Study Time vs. Final Grade (G3)')
plt.xlabel('Study Time (hours/week)')
plt.ylabel('Final Grade (G3)')
plt.grid(linestyle='--', alpha=0.7)
plt.show()

# Plot 3: Bar Chart (Average Scores by Gender)
plt.figure(figsize=(8, 5))
gender_avg_scores.plot(kind='bar', color=['blue', 'orange'])
plt.title('Average Final Grades by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Final Grade (G3)')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Step 6: Documentation
# I have documented each step as comments in the code.
# This ensures clarity and provides an explanation for the logic and operations used.
```
